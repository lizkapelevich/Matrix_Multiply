```python
a = np.linspace(0, 1, 1000)
```

```python
t = 10*a
```

```python
f = 0.5
A = 1
omega = 2*np.pi*f
x1 = A*np.cos(omega*t)
```

```python
plt.rcParams.update({'font.size': 18})
plt.figure(figsize=(12,7))
plt.plot(t, x1)
plt.title("Time vs. Position")
plt.xlabel("Time")
plt.ylabel("Position")
```

```python
B = 1
x2 = A*np.sin(omega*t)+B*np.cos(omega*t)
```

```python
x3 = np.sqrt(2)*np.cos(omega*t-np.pi/4)
```

```python
plt.rcParams.update({'font.size': 18})
plt.figure(figsize=(12,7))
plt.plot(t, x2, label = "Second Wave")
plt.plot(t, x1, label = "First Wave")
plt.plot(t, x3, '--', label = "Third Wave")
plt.title("Comparing Waves with Different Times & Positions")
plt.xlabel("Time")
plt.ylabel("Position")
plt.legend()
```

### Find a function that has a chirp behavior:
#### (1) As time evolves the frequency increases
#### (2) As time evolves the amplitude also increases
#### (3) Combine them

### Create this signal and plot it.

```python
x4 = np.sin((2*np.pi*f)*t**2);
plt.plot(t,x4)
plt.title("Frequency Increasing Over Time")
plt.xlabel("Time")
plt.ylabel("Position")
```

```python
x5 = (A*t)**2*(np.sin((2*np.pi)*t))
plt.plot(t,x5)
plt.title("Amplitude Increasing Over Time")
plt.xlabel("Time")
plt.ylabel("Position")
```

```python
x6 = (A*t)**2*(np.sin((2*(np.pi)*t**2)))
plt.plot(t,x6)
plt.title("Combination of Amplitude and Frequency Increasing Over Time")
plt.xlabel("Time")
plt.ylabel("Position")
```
